```python
#네이버이미지 크롤링
```

```python
import selenium,time,os,zipfile
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from urllib.request import urlretrieve
from tqdm import tqdm
```

```python
# 키워드 입력
keyword = input("수집할 이미지는 >>>")
if keyword == '':
    keyword='벗꽃'
    
print('[접속중.....]')
driver = webdriver.Chrome()
driver.implicitly_wait(15)
url = f'https://search.naver.com/search.naver?where=image&sm=tab_jum&query={keyword}'
driver.get(url)

# 화면 스크롤
print('[화면 스크롤.....]')
body = driver.find_element_by_css_selector('body')
for i in range(5):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)

links = []
imgs = driver.find_elements_by_css_selector('img._image')
print(len(imgs))

# 이미지 링크 추출
print('[이미지 링크 추출 중.....]')
for img in imgs:
    link = img.get_attribute('src')
    # print(link)
    if 'https' in link:
        links.append(link)
# print(links)
print('[정보 수집 완료 .....]')

# 폴더 생성
print('[폴더 생성 중.....]')
if not os.path.isdir(f'./{keyword}'):
    os.mkdir(f'./{keyword}')

# 이미지 다운로드   
print('[이미지 다운로드 중.....]')
for index,link in tqdm(enumerate(links),total=len(links)):
    start = link.rfind('.')
    end = link.rfind('&')
    file_type = link[start:end]
    # print(file_type)
    filename=f'./{keyword}/{keyword}{index:03d}{file_type}'
    # print(filename)
    urlretrieve(link,filename)
print('[이미지 다운로드 완료.....]')

#파일압축
print('[이미지 압축 중.....]')
zip_file = zipfile.ZipFile(f'./{keyword}.zip','w')
for image in os.listdir(f'./{keyword}'):
    # print(image)
    zip_file.write(f'./{keyword}/{image}',compress_type=zipfile.ZIP_DEFLATED)
zip_file.close()
print('[이미지 압축 완료.....]')
```
