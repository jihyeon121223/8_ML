```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://finance.naver.com/marketindex/exchangeList.naver'
res=requests.get(url).text
soup=BeautifulSoup(res,'lxml')
result=soup.select('td.tit a')
result
name=[]
name1=[]
price=[]
for item in result:
    #name.append(item.string.strip()) 
    v=item.string.strip().split() #나라 이름 사이 공백 제거
    name.append(v[0].strip())   #strip 공백제거
    name1.append(v[1].strip())
    
result=soup.select('td.sale')
for item in result:
    price.append(float(item.string.replace(',','')))

list(zip(name,name1,price))
```

```python
#Bugs차트 앨범,가수 뽑기
```

```python
url='https://music.bugs.co.kr/chart'
res=requests.get(url).text
res
soup=BeautifulSoup(res,'html.parser')

title_list=[]
title=soup.select('p.title a')
for item in title:
    #print(item.text)
    title_list.append(item.text)
    
artist_list=[]
artist=soup.select('p.artist')
for item in artist:
    artist_list.append(item.text.replace('\n',''))
    
result=list(zip(title_list,artist_list))
print(result)
```

```python
from urllib.request import urlopen
import requests
url='https://news.naver.com/main/read.naver?mode=LSD&mid=shm&sid1=104&oid=001&aid=0013045122'
res=requests.get(url).read()
soup=BeautifulSoup(res,'html.parser')
news=soup.select('ul.list_text li a')
for item in news:
    #title=item.text
    title=item['title']
    print(title)
    url=item['href']
    res=urlopen(url)
    soup=BeautifulSoup(res,'html.parser')
    #print(soup)
    contents=soup.select_one('#main_content')
    print(contents.text.strip())
```

```python
#네이버 영화 평점 뽑기
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
movie_list=[]
for i in range(1,11):
    url=f'https://movie.naver.com/movie/point/af/list.naver?page={i}'
    res=urlopen(url)
    soup=BeautifulSoup(res,'html.parser')
    table=soup.select('table.list_text li a')
    for i,r in enumerate(table.select('tbody > tr')):
        #print(i)
        #print(r)
        for j,c in enumerate(r.select('td')):
            #print(j)
            #print(c)
            if j ==1:
                #print(c)
                title=c.select_one('a').text
                print(title)
                jumsu=c.select_one('em').text
                #print(jumsu)
                #print(c.text)
                cont=c.text.replace(title,''),replace('신고','')
                cont=re.sub('별점 - 총 10점 중[0-9]{1,2}','',cont)
                cont=cont.strip()
                #print(cont)
                #dic={'제목':title,'평점':jumsu,'감상평':cont}
                lis=(title,jumsu,cont)
                movie_list.append(lis)
print(movie_list)

import json
with open('movielist.json','w') as f:
    json.dump(movie_list,f)
```

```python
import sqlite3
con=sqlite3.connect('movie.db')
cur=con.cursor()
cur.execute('''
            create table if not exists movie
            (title text, point int, cont text)
            ''')
con.commit()
```

```python
cur.executemany('''
            insert into movie values(?,?,?),movie_list
            ''')
con.commit()
cur.close()
```
