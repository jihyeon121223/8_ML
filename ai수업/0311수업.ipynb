```python
driver=webdriver.Chrome()
driver.get('https://www.starbucks.co.kr/store/store_map.do')
time.sleep(5)
try:
    w=driver.switch_to.alert()
    #print(w.text)
    w.accept()
    #w.dismiss()
except:
    pass
time.sleep(5)
loca=driver.find_element_by_class_name('loca_search')
loca.click()
time.sleep(3)
sido=driver.find_elements_by_class_name('set_sido_cd_btn')
sido[5].click()
time.sleep(3)
gugun=driver.find_elements_by_class_name('set_gugun_cd_btn')
gugun[4].click()
time.sleep(10)
source=driver.page_source
soup=BeautifulSoup(source,'html.parser')
li_list=soup.select('ul.quickSearchResultBoxSidoGugun li')
#li_list
for item in li_list:
    print(item.select_one('strong').text)
    print(item.select_one('p').text)
    print('==============')
```
